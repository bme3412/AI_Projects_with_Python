```python
from langchain.agents import create_pandas_dataframe_agent
from langchain.chat_models import ChatOpenAI
from langchain.agents.agent_types import AgentType
from langchain.llms import OpenAI
import pandas as pd
```

```python
paris_airbnb = pd.read_csv('paris_listings.csv', index_col=0)
```

```python
paris_airbnb = paris_airbnb[paris_airbnb['room_type']=='Entire home/apt']
```

```python
agent = create_pandas_dataframe_agent(
    ChatOpenAI(openai_api_key = authentication_key,temperature=0, model="gpt-3.5-turbo-0613"),
    paris_airbnb,
    verbose=True,
    agent_type=AgentType.OPENAI_FUNCTIONS,
)
```

```python
query = "Which listings has the best value for money? Please list the top 10 and save to a dataframe"
query = query + " using tool python_repl_ast"
agent.run(query)


#agent.run("Which listing has the best value for money?")
```
